#### **Crew AI has several offerings**

1. CrewAI Enterprise: A multi-agent platform for deploying, running and monitoring Agentic AI
2. Crew UI studio: A no-code/lo-code product for creating multi-agent solutions
3. CrewAI open source framework: Orchestrate high performing AI agents with ease and scale 


#### **The crewai comes with two flavours**

1. CrewAI Crews: Autonomous solutions with AI teams of agents with different roles
2. CrewAI Flows: Structured Automations by dividing complex tasks into precise workflows 

**Choose crews when: You need an autonoumous problem-solving, creative collaboratiob, or exploratory tasks.<br> Choose Flows when: You require outcomes, auditability or precise control over execution**


#### **Core Concepts**

1. **Agent**: An autonomous unit, with an LLM, a role, a goal, a backstory, memory, tools
2. **Task**: A specific assignment to be carried out, with a description, expected output, agent
3. **Crew**: A team of Agents and Tasks, either:<br>
        a. Sequential: Run tasks in order they are defined <br>
        b. Heirarchical: Use a manager LLM to assign

#### Memory in CrewAI (stock pickr)

CrewAI provides multiple types of memory that allow agents to recall, reuse, and build upon past interactions.

#### **Short-Term Memory**

* Temporarily stores recent interactions and outcomes.
* Uses **RAG (Retrieval-Augmented Generation)** so agents can access relevant information during the **current execution**.
* Example: A researcher agent remembers the last search results during the same run.

#### **Long-Term Memory**

* Preserves valuable insights and learnings across multiple runs.
* Builds **knowledge over time**, enabling agents to become smarter with continued usage.
* Example: An analyst agent recalls past financial trends discovered in earlier projects.

#### **Entity Memory**

* Stores structured information about **people, places, and concepts** encountered.
* Facilitates **relationship mapping** and deeper contextual understanding.
* Uses **RAG** for storing and retrieving entity-specific information.
* Example: Remembering details about Apple Inc. across different tasks.

#### **Contextual Memory**

* Maintains the **overall context** by combining short-term, long-term, and entity memory.
* Ensures smoother, more coherent interactions across multiple agents and tasks.
* Example: A judge agent evaluates debates with awareness of prior arguments, insights, and entities involved.
